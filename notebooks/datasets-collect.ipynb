
### Import libraries

```python
import requests
import json
import pandas as pd
import os
from dotenv import load_dotenv 
```

### Import Bearer token from .env file

```python
load_dotenv()
TOKEN = os.getenv('TOKEN')
```

### Query Parameter & Excecute request

```python
API_URL ='https://api.github.com/search/'
TYPE = 'repositories'
QUERY_STRING = 'Data scientist+data analytic'
STARS = '>=10'
DATE_FROM = '2000-01-01'
DATE_TO = '2023-12-31'
LANGUAGE = 'Jupyter Notebook'
PAGE = 1

requestUrl =F'{API_URL}{TYPE}?q={QUERY_STRING}+stars:{STARS}+created:{DATE_FROM}..{DATE_TO}+language:"{LANGUAGE}"&page={PAGE}&per_page=100'
headers = {"Authorization": "Bearer {}".format(TOKEN)}
response = requests.get(requestUrl, headers=headers)
print(requestUrl)
print(response)
```
